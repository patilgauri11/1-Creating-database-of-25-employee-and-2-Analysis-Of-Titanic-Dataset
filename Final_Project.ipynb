1.Make a script to take 25 employee data id,name,gender,age,city,sal,exp,working for the company years? dynamic--input()
Make 2 null values  in categories city,gender and salary ,exp all in different rows
Random Password Generator  :Password must have capital,small,number,symbol,leng 8 plus generate  25 password for above employee?


```python
import random
db={}
sum=0
for i in range(25):
    id_no=int(input("id:"))
    name=input("name: ")
    gender=input("gender:")
    age=int(input("age:"))
    city=input("city:")
    sal=float(input("salary:"))
    years=int(input("years:"))
    
    chars ="abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ1234567890!@#$%^&*()_+}{"
    length = int(input("Enter password length "))
    password=""
    for i in range(length+1):
        password += random.choice(chars)
    print(password)
    
    l=[]
    l.append(id_no)
    l.append(name)
    l.append(gender)
    l.append(age)
    l.append(city)
    l.append(sal)
    l.append(years)
    l.append(password)
    db[id_no]=l
    

db
```

```python
import pandas as pd
df=pd.DataFrame(db)
df
df1 = df.T
df1
```

```python
df1.rename(columns = {0:'id'}, inplace = True)
df1.rename(columns = {1:'Name'}, inplace = True)
df1.rename(columns = {2:'gender'}, inplace = True)
df1.rename(columns = {3:'age'}, inplace = True)
df1.rename(columns = {4:'city'}, inplace = True)
df1.rename(columns = {5:'salary'}, inplace = True)
df1.rename(columns = {6:'Experience'}, inplace = True)
df1.rename(columns = {7:'password'}, inplace = True)
df1
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rc("font", size = 14)
import seaborn as sns
sns.set(style = "white")
sns.set(style = "whitegrid", color_codes = True)
data = pd.read_csv("Titanic_train_project2.csv")    # Reading titanic csv file
```

```python
data.isnull().sum()  # cheking is there a null values in data
```

```python
# how age variable looks like
a = data["Age"].hist(bins=15, color='teal', alpha=0.8)
a.set(xlabel='Age', ylabel='Count')
plt.savefig('Age Count.png')
plt.show()
```

Since "Age" is (right) skewed, using the mean might give us biased results by filling in ages. 
To deal with this, we'll use the median to impute the missing values. 


```python
data['Age'].median(skipna=True)   # median age is 28
```

Missing values for Cabin
787 of records are missing, which means that imputing information and using this variable
for prediction is probably not wise. We'll ignore this variable in our model.

# Embarked missing values

Embarked Missing values
There are only 2 missing values for "Embarked", 
so we can just impute with the port where most people boarded.

```python

#sns.barplot('Sex', 'Survived', data = data, color = "Blue")
sns.countplot(x='Embarked', data=data, palette='Set2')
plt.savefig('Embarked count.png')
plt.show()
```

```python
data['Age'].fillna(28, inplace=True)  # Median age  is 28
data['Embarked'].fillna('S', inplace=True)  # most common are S
data.drop('Cabin', axis=1, inplace=True)  #Ignoring this variable in our model because it has almost 687 records are missing
data.drop('Name', axis=1, inplace=True)   # delete useless column
data
```
